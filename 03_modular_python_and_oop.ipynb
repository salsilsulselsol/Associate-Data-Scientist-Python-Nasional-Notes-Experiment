## 1. Functions and Help
Functions are reusable blocks of code designed to perform specific tasks. They promote modularity and abstraction.

```python
# 1.1 Defining and Calling Functions
def calculate_area(radius, pi=3.14):
    """Calculates the area of a circle. Returns the result."""
    return pi * (radius ** 2)

# Calling with arguments
print(f"Area: {calculate_area(10)}")
```

```python
# 1.2 Lambda Functions (Anonymous)
# Useful for short, one-liner operations
square = lambda x: x ** 2
print(f"Square of 4: {square(4)}")
```

```python
# 1.3 Variable-Length Arguments (*args)
def sum_all(*args):
    return sum(args)

print(f"Sum: {sum_all(10, 20, 30, 40)}")
```

```python
# 1.4 Getting Help
# Use help() to see documentation
help(calculate_area)
```

## 4. Error Handling
Robust code must handle exceptions (runtime errors) gracefully using try-except blocks.

```python
try:
    value = 10 / 0
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
finally:
    print("Execution completed.")
```

## 5. Object-Oriented Programming (OOP)
OOP uses 'Classes' as blueprints for 'Objects'. This is how modern AI libraries are built.

```python
# 5.1 Classes and Inheritance
class Animal:
    def speak(self):
        return "Animal sound"

class Dog(Animal): # Inheritance
    def __init__(self, name):
        self.name = name

    def speak(self): # Polymorphism (Method Overriding)
        return f"{self.name} says Woof!"

my_dog = Dog("Buddy")
print(my_dog.speak())
```

```python
# 5.2 Encapsulation (Protecting Data)
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance # Private attribute

    def get_balance(self):
        return self.__balance
```

```python
# 5.3 Abstraction
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def move(self):
        pass

class Car(Vehicle):
    def move(self):
        return "The car drives on roads."
```
