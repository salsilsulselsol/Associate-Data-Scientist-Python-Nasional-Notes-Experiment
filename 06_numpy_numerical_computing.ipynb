## 1. Introduction to NumPy
NumPy (Numerical Python) is optimized in C, making it significantly faster than standard Python lists. It is the backbone for Pandas, Scikit-learn, and TensorFlow.

### 1.1 Creating Arrays
Arrays can be created from lists or using built-in functions for specific distributions.

```python
import numpy as np

# Creating from a list
arr_2d = np.array([[1, 2], [3, 4]])

# Specialized creation functions
zeros_arr = np.zeros((3, 3))              # 3x3 array of 0.0
ones_arr = np.ones((2, 3))               # 2x3 array of 1.0
random_arr = np.random.rand(2, 2)        # Random values [0, 1)
range_arr = np.arange(0, 10, 2)          # [0, 2, 4, 6, 8]
linspace_arr = np.linspace(0, 1, 5)      # 5 values evenly spaced from 0 to 1

# Array Attributes
print(f"Shape: {arr_2d.shape}")
print(f"Size: {arr_2d.size}")
print(f"Data Type: {arr_2d.dtype}")
```

## 2. Array Operations
Operations in NumPy are performed **element-wise**, allowing for concise and fast calculations.

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Arithmetic
print(f"Addition: {a + b}")
print(f"Multiplication: {a * b}")

# Indexing and Slicing
arr = np.array([10, 20, 30, 40, 50])
print(f"Slice (index 1 to 3): {arr[1:4]}")
print(f"Last element: {arr[-1]}")
```

## 3. Shape Manipulation
Changing data dimensions is crucial for preparing inputs for Deep Learning models.

```python
# Reshaping
arr_1d = np.arange(6)
reshaped = arr_1d.reshape(2, 3)

# Stacking and Concatenation
v_stack = np.vstack((a, b))    # Vertical stack
h_stack = np.hstack((a, b))    # Horizontal stack
concat = np.concatenate((a, b))

print(f"Reshaped Matrix:\n{reshaped}")
```

## 4. Mathematical and Statistical Functions
NumPy provides powerful tools for descriptive statistics and matrix operations.

```python
data = np.array([[1, 2], [3, 4]])

# Statistics
print(f"Mean: {np.mean(data)}")
print(f"Median: {np.median(data)}")
print(f"Standard Deviation: {np.std(data)}")

# Linear Algebra (Matrix Multiplication)
# Essential for calculating transformations and weights
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[5, 6], [7, 8]])
dot_product = np.dot(matrix_a, matrix_b)

print(f"Dot Product result:\n{dot_product}")
```

## 5. Practical Use Case: Axis-based Calculations
In Data Science, we often need to calculate metrics across rows (axis 1) or columns (axis 0).

```python
# Simulating 5 students with scores in 3 subjects
np.random.seed(42)
scores = np.random.rand(5, 3) * 100

# Median score for each subject (Column-wise)
subject_medians = np.median(scores, axis=0)

# Average score for each student (Row-wise)
student_averages = np.mean(scores, axis=1)

print(f"Subject Medians: {subject_medians}")
print(f"Student Averages: {student_averages}")
```
