## 1. Syntax, Comments, and Variables
Python uses indentation to define code blocks and does not require explicit variable declarations (dynamic typing).

```python
# 1.1 Comments
# This is a single-line comment
"""
This is a multi-line comment or docstring.
Often used for documentation.
"""

# 1.2 Variables and Dynamic Typing
# Variables store data that can change during runtime.
age = 21                # Integer
height = 181            # Float
name = "Faisal"         # String
is_student = True       # Boolean

# Output and type checking
print(f"Name: {name}, Type: {type(name)}")
print(f"Age: {age}, Type: {type(age)}")
```

## 2. Operators
Operators allow us to perform mathematical and logical operations.

```python
# 2.1 Arithmetic Operators
a, b = 15, 4
print(f"Addition: {a + b}")
print(f"Modulus (Remainder): {a % b}") # Useful for even/odd checks

# 2.2 Comparison Operators
print(f"Is 10 equal to 10? {10 == 10}")
print(f"Is 7 not equal to 8? {7 != 8}")

# 2.3 Logical Operators
# Used to combine boolean expressions
x_valid = True
y_valid = False

print(f"AND Logic: {x_valid and y_valid}") # True if both are true
print(f"OR Logic: {x_valid or y_valid}")   # True if at least one is true
print(f"NOT Logic: {not x_valid}")         # Reverses the boolean
```

## 3. Conditional Statements
Conditionals allow the program to make decisions based on specific criteria.

```python
# 3.1 If-Elif-Else Structure
score = 85

if score >= 90:
    grade = "A"
    feedback = "Excellent work!"
elif score >= 80:
    grade = "B"
    feedback = "Very good!"
elif score >= 70:
    grade = "C"
    feedback = "Keep improving."
else:
    grade = "F"
    feedback = "Needs more effort."

print(f"Grade: {grade}\nFeedback: {feedback}")

# 3.2 Nested Conditionals
age = 20
has_license = True

if age >= 18:
    if has_license:
        print("Eligible to drive.")
    else:
        print("Need a license to drive.")
else:
    print("Too young to drive.")
```

## 4. Practical Applications and Best Practices
Applying logic to real-world scenarios and avoiding common errors.

```python
# 4.1 Even or Odd Determination
number = 7
if number % 2 == 0:
    print(f"{number} is Even.")
else:
    print(f"{number} is Odd.")

# 4.2 Best Practice: Descriptive Variable Names
# Avoid: a = 5, b = 10
# Use:
min_threshold = 5
current_value = 10

if current_value > min_threshold:
    print("Condition met.")
```
