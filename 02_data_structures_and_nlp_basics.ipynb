## 1. Lists
Lists are ordered, mutable, and heterogeneous collections. They allow us to store multiple items in a single variable.

### 1.1 Creation and Indexing
Python uses zero-based indexing. Negative indexing allows access from the end of the list.

```python
# Creating lists
fruits = ['apple', 'banana', 'cherry']
mixed_list = [1, "Data Science", True, 3.14]

# Indexing
print(f"First element: {fruits[0]}")     # apple
print(f"Last element: {fruits[-1]}")      # cherry

# Modifying elements (Mutable)
fruits[1] = 'blueberry'
print(f"Modified list: {fruits}")
```

### 1.2 List Methods
Methods to add, remove, and extend list data.

```python
# Adding elements
fruits.append('dragonfruit')        # Adds to the end
fruits.insert(1, 'avocado')         # Inserts at index 1
fruits.extend(['grape', 'kiwi'])    # Appends multiple elements

# Removing elements
fruits.remove('cherry')             # Removes by value
popped_item = fruits.pop(0)         # Removes by index and returns it
del fruits[1]                       # Deletes by index

print(f"Final list: {fruits}")
```

## 2. Loops: Automating Repetitive Tasks
Loops are essential for processing datasets, performing batch operations, and implementing algorithms.

```python
# 2.1 For Loop with Range
# range(start, stop) generates numbers from start to stop-1
for i in range(5):
    print(f"Iteration: {i}")

# 2.2 Iterating through a List with Enumerate
# enumerate() provides both index and value
for index, fruit in enumerate(fruits):
    print(f"Index {index}: {fruit}")

# 2.3 While Loop
# Runs as long as the condition is True
counter = 5
while counter > 0:
    print(f"Countdown: {counter}")
    counter -= 1
```

## 3. Strings
Strings are immutable sequences of characters. Mastering string methods is vital for Natural Language Processing (NLP).

```python
# 3.1 Slicing and Formatting
message = "  Learning Python for AI  "

# Slicing: [start:stop]
print(f"Slice (0-8): {message[2:10]}")

# Formatting with f-strings
topic = "Deep Learning"
print(f"Focus area: {topic}")

# 3.2 Common String Methods
clean_message = message.strip().upper() # Removes whitespace and uppercases
replaced_text = clean_message.replace("PYTHON", "LLM")

print(f"Cleaned: '{clean_message}'")
print(f"Replaced: '{replaced_text}'")
```

## 4. Dictionaries
Dictionaries store data in key-value pairs. They are highly efficient for data retrieval and configuration management.

```python
# 4.1 Creation and Access
student = {
    "name": "Faisal",
    "major": "Computer Science",
    "gpa": 3.8,
    "skills": ["Python", "Computer Vision"]
}

# Accessing values
print(f"Student Name: {student['name']}")
# Safe access using .get() to avoid KeyError
print(f"Age: {student.get('age', 'Not Specified')}")

# 4.2 Nested Dictionaries
university = {
    "department": "IT",
    "students": [
        {"name": "Ani", "year": 2},
        {"name": "Budi", "year": 3}
    ]
}
```

## 5. Comprehensions
Comprehensions provide a concise syntax for creating new collections from existing ones.

```python
# 5.1 List Comprehension
numbers = [1, 2, 3, 4, 5, 6]
# Create squares for even numbers only
even_squares = [x**2 for x in numbers if x % 2 == 0]
print(f"Even Squares: {even_squares}")

# 5.2 Dictionary Comprehension
# Creating a mapping of numbers to their cubes
cube_map = {x: x**3 for x in range(5)}
print(f"Cube Map: {cube_map}")

# 5.3 Nested List Comprehension (Flattening a Matrix)
matrix = [[1, 2], [3, 4]]
flattened = [element for row in matrix for element in row]
print(f"Flattened Matrix: {flattened}")
```
