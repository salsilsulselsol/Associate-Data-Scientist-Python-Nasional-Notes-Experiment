## 1. Linear Algebra
Linear algebra provides the mathematical framework for algorithms that process multi-dimensional data.

### 1.1 Matrices and Operations
Matrices represent data structures (e.g., movie ratings in recommendation systems).

```python
import numpy as np

# Creating a 2x3 matrix
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8, 9], [10, 11, 12]])
```

```python
# 1.1.1 Matrix Addition (e.g., Merging sensor data in IoT)
matrix_sum = A + B
```

```python
# 1.1.2 Matrix Multiplication (Dot Product)
# Essential for image transformations and Neural Network layers
# Requirement: columns of A must match rows of C
C = np.array([[1, 2], [4, 5], [7, 8]])
matrix_product = np.dot(A, C)
```

```python
# 1.1.3 Transpose (Reshaping data for Neural Networks)
matrix_transpose = A.T

print(f"Product result:\n{matrix_product}")
```

### 1.2 Matrix Properties
Determining stability and solvability of linear systems.

```python
# 1.2.1 Determinant (Stability Analysis)
M = np.array([[4, 2], [3, 1]])
det_M = np.linalg.det(M) # (4*1)-(2*3) = -2
```

```python
# 1.2.2 Inverse (Solving linear equations in Physics)
M_inv = np.linalg.inv(M)
```

```python
# 1.2.3 Rank (Dimensionality reduction in Bioinformatics)
rank_M = np.linalg.matrix_rank(M)
```

```python
# 1.2.4 Eigenvalues and Eigenvectors (Vibration Analysis/PCA)
eigenvalues, eigenvectors = np.linalg.eig(M)
```

## 2. Probability Theory
Probability is the core of predictive modeling and evaluating data reliability.

### 2.1 Basic Probability and Independence

```python
# 2.1.1 Simulated Die Roll (Law of Large Numbers)
np.random.seed(0)
throws = np.random.randint(1, 7, size=10000)
prob_even = np.sum(throws % 2 == 0) / len(throws)
print(f"Simulated probability of even number: {prob_even}")
```

```python
# 2.1.2 Covariance (Measuring linear dependence)
X = np.random.normal(0, 1, 1000)
Y = np.random.normal(0, 1, 1000)
cov_matrix = np.cov(X, Y)
print(f"Covariance between X and Y: {cov_matrix[0, 1]}")
```

### 2.2 Common Probability Distributions
Visualizing data spread is crucial for feature engineering.

```python
import matplotlib.pyplot as plt

# Prepare figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Normal Distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)
axes[0].hist(normal_data, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title("Normal Distribution")

# Binomial Distribution
binomial_data = np.random.binomial(n=10, p=0.5, size=1000)
axes[1].hist(binomial_data, bins=30, color='lightgreen', edgecolor='black')
axes[1].set_title("Binomial Distribution")

# Poisson Distribution
poisson_data = np.random.poisson(lam=3, size=1000)
axes[2].hist(poisson_data, bins=30, color='salmon', edgecolor='black')
axes[2].set_title("Poisson Distribution")

plt.tight_layout()
plt.show()
```

## 3. Calculus
Calculus provides tools for analyzing and optimizing algorithms, especially in minimizing error functions.

### 3.1 Types of Functions

```python
# Linear, Quadratic, and Polynomial
x = np.linspace(-10, 10, 400)
y_linear = 2 * x + 1
y_quadratic = x**2 - 4*x + 3
y_poly = 0.1 * x**3 - 0.5 * x**2 + 2 * x - 1

# Side-by-side plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Linear Plot
axes[0].plot(x, y_linear, color='blue')
axes[0].set_title('Linear: 2x + 1')
axes[0].grid(True)

# Quadratic Plot
axes[1].plot(x, y_quadratic, color='green')
axes[1].set_title('Quadratic: x^2 - 4x + 3')
axes[1].grid(True)

# Polynomial Plot
axes[2].plot(x, y_poly, color='red')
axes[2].set_title('Polynomial: 0.1x^3 - 0.5x^2 + 2x - 1')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

### 3.2 Derivatives and Gradient Descent
Derivatives measure the rate of change. Gradient Descent finds the minimum of a cost function.

```python
# Cost Function f(x) = x^2 - 4x + 3
def f(x): return x**2 - 4*x + 3
def df(x): return 2*x - 4 # First derivative

# Optimization via Gradient Descent
def gradient_descent(init_x, lr=0.1, iters=50):
    x = init_x
    history = [x]
    for _ in range(iters):
        x -= lr * df(x)
        history.append(x)
    return x, history

# Run optimization
best_x, path = gradient_descent(init_x=0.0)
print(f"Minimum found at x = {best_x}")

# Gradient Descent Visualization
x_range = np.linspace(-1, 5, 100)
plt.figure(figsize=(10, 6))
plt.plot(x_range, f(x_range), label='f(x) = x^2 - 4x + 3', color='blue')

# Plot optimization steps
path_y = [f(p) for p in path]
plt.scatter(path, path_y, color='red', label='Optimization Path', s=10, alpha=0.5)
plt.plot(path, path_y, color='red', linestyle='--', alpha=0.3)

# Plot final minimum point (Star)
plt.scatter(best_x, f(best_x), color='gold', marker='*', s=200, label='Minimum found', zorder=5)

plt.title("Gradient Descent Optimization (Converged)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

### 3.3 Integration
Represents the area under the curve, used in calculating probabilities (PDF).

```python
# Integral Visualization (Area Under Curve)
def f_int(x): return x**2

x_full = np.linspace(0, 3, 400)
y_full = f_int(x_full)

plt.figure(figsize=(8, 5))
plt.plot(x_full, y_full, 'r', linewidth=2, label='f(x) = x^2')

# Define integration limits
a, b = 0.5, 2.5
x_fill = np.linspace(a, b, 100)
y_fill = f_int(x_fill)

# Shade the area under the curve
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.4, label=f'Integral area (a={a} to b={b})')

# Add vertical boundary lines
plt.vlines([a, b], 0, [f_int(a), f_int(b)], colors='gray', linestyles='--')

plt.title("Integral Visualization as Area Under the Curve", fontsize=14)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.text(1.5, 1.0, r'$\int_a^b f(x)dx$', fontsize=16, horizontalalignment='center')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
