## 1. Linear Algebra
Linear Algebra allows us to represent data as multi-dimensional arrays (tensors). The Dot Product is the fundamental operation behind Neural Networks.

```python
import numpy as np

# 1.1 Vector and Matrix Representation
vector_a = np.array([1, 2, 3])
matrix_a = np.array([[1, 2], [3, 4]])
```

```python
# 1.2 Dot Product (Matrix Multiplication)
# Crucial for calculating weighted sums in neurons
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

dot_result = np.dot(A, B)
print(f"Matrix Dot Product:\n{dot_result}")
```

```python
# 1.3 Transpose and Inverse
# Transpose is used to align matrix dimensions for operations
matrix_t = A.T
print(f"Transpose of A:\n{matrix_t}")
```

## 2. Probability and Statistics
Probability helps us quantify uncertainty. Statistics allows us to understand the distribution and variance of our features.

```python
# 2.1 Descriptive Statistics
data = np.array([12, 15, 12, 18, 20, 25, 30])

mean_val = np.mean(data)
std_dev = np.std(data)

print(f"Mean: {mean_val}")
print(f"Standard Deviation (Spread of data): {std_dev}")
```

## 3. Calculus
Calculus is the engine of optimization. We use derivatives to find the direction that minimizes the error in our models (Gradient Descent).

```python
import sympy as sp

# 3.1 Symbolic Differentiation
x = sp.Symbol('x')
y = sp.Symbol('y')

# Defining a cost function: f(x) = x^2 + 5x + 10
f = x**2 + 5*x + 10

# Calculating the derivative (Gradient)
derivative = sp.diff(f, x)
print(f"Function: {f}")
print(f"Derivative (Slope): {derivative}")
```

```python
# 3.2 Partial Derivatives
# Important for models with multiple weights
g = x**2 + y**3
partial_x = sp.diff(g, x)
partial_y = sp.diff(g, y)

print(f"Partial derivative w.r.t x: {partial_x}")
print(f"Partial derivative w.r.t y: {partial_y}")
```
