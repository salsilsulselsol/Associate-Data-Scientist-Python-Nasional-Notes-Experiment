## 1. Creating, Reading, and Writing Data
Pandas provides two main structures: **Series** (1D) and **DataFrame** (2D).

```python
import pandas as pd
import numpy as np

# 1.1 Creating a DataFrame from a Dictionary
data = {
    'Product': ['Laptop', 'Smartphone', 'Tablet'],
    'Price': [1200, 800, 400],
    'Stock': [15, 50, 20]
}
df = pd.DataFrame(data, index=['Item_1', 'Item_2', 'Item_3'])
```

```python
# 1.2 Loading and Saving Data
# Reading a CSV (Example syntax)
# df = pd.read_csv('data.csv', encoding='utf-8')

# Checking the first few rows and dimensions
print(df.head(2))
print(f"Dataset Shape: {df.shape}")

# Saving to CSV without index
df.to_csv('output_products.csv', index=False)
```

## 2. Indexing, Selection, and Assignment
Accessing specific data points using labels (`.loc`) or positions (`.iloc`).

```python
# 2.1 Label-based selection (.loc)
print(df.loc['Item_1', 'Price'])
```

```python
# 2.2 Position-based selection (.iloc)
# Getting the first row, first two columns
print(df.iloc[0, 0:2])
```

```python
# 2.3 Conditional Selection (Filtering)
# Crucial for subsetting data for specific ML tasks
expensive_items = df[df['Price'] > 500]

# Combining conditions
filtered_data = df[(df['Price'] > 300) & (df['Stock'] < 30)]
```

```python
# 2.4 Assignment
df['Discounted_Price'] = df['Price'] * 0.9
```

## 3. Summary Functions and Maps
Transforming data systematically using built-in functions or custom logic.

```python
# 3.1 Descriptive Statistics
print(df.describe())
print(f"Mean Price: {df['Price'].mean()}")
```

```python
# 3.2 Value Counts (Useful for categorical features)
print(df['Product'].value_counts())
```

```python
# 3.3 Mapping and Applying Functions
# Simple mapping using lambda
df['Price_Category'] = df['Price'].map(lambda x: 'High' if x > 1000 else 'Budget')

# Complex transformation using apply
def categorize_stock(qty):
    if qty > 40: return 'Abundant'
    return 'Limited'

df['Stock_Status'] = df['Stock'].apply(categorize_stock)
```

## 4. Grouping and Sorting
Aggregating data to find patterns across categories.

```python
# 4.1 Basic Grouping
# Calculating average price per category
avg_price = df.groupby('Price_Category')['Price'].mean()
```

```python
# 4.2 Multi-aggregation
group_stats = df.groupby('Price_Category')['Stock'].agg(['sum', 'mean', 'count'])
```

```python
# 4.3 Sorting
# Sorting by Price descending
sorted_df = df.sort_values(by='Price', ascending=False)
```

## 5. Data Types and Missing Values
Identifying and treating 'NaN' values to prevent model bias.

```python
# Checking data types
print(df.dtypes)

# Identifying missing values
df.isnull().sum()

# Strategies for handling missing data
df.dropna()                          # Removing rows
df.fillna(0)                         # Filling with zero
df['Price'].fillna(df['Price'].mean()) # Imputation with mean

# Replacing placeholders (e.g., '?' or 'NA')
df.replace(['?', '-'], np.nan, inplace=True)
```

## 6. Renaming and Combining
Merging datasets from different sources into a single coherent structure.

```python
# 6.1 Renaming Columns
df.rename(columns={'Price': 'Unit_Price', 'Stock': 'Inventory'}, inplace=True)
```

```python
# 6.2 Combining DataFrames
df_extra = pd.DataFrame({'Product': ['Monitor'], 'Unit_Price': [200], 'Inventory': [10]})

# Concatenation (Row-wise)
combined_df = pd.concat([df, df_extra], axis=0)

# Merging (SQL-like join)
# result = pd.merge(df1, df2, on='Key_Column', how='inner')
```
