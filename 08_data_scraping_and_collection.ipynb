# Fundamentals of Data

Data is a collection of facts, information, or numbers arranged in a format that can be processed by computers or humans. It represents various phenomena or events that are observed or measured.

### Data Forms
Data exists in several forms:
* **Text**: Documents, emails, or database entries.
* **Numbers**: Measurements, statistics, or financial information.
* **Images**: Photos, graphics, or medical imagery.
* **Audio & Video**: Voice recordings, music, or visual signals.

### Data Classification
Data is categorized based on its characteristics to determine the appropriate analysis method:

* **Qualitative Data**: Descriptive information that is difficult to measure numerically.
    * **Nominal**: Variables with no inherent order (e.g., Gender, Race).
    * **Ordinal**: Variables with an ordered series or ranking (e.g., Performance, Blood Group).
    * **Binary**: Variables with only two options (e.g., Pass/Fail, Yes/No).
* **Quantitative Data**: Information that can be measured and counted numerically.
    * **Discrete**: Data based on counts with finite values (e.g., number of students).
    * **Continuous**: Data measured on a scale that can be subdivided (e.g., Height, Weight).

### Components of Data
A structured dataset consists of three essential parts:

* **Features**: Also known as Columns, Fields, or Attributes. These are the variables used for analysis.
* **Labels**: Also known as Classes or Groups. This is the target variable in predictive modeling.
* **Examples**: Also known as Samples, Subjects, or Rows. These represent individual observations.

## Section 2: Data Collection Techniques

### Sources of Data
* **Primary Data**: Obtained directly from the source or object studied. Highly accurate but resource-intensive.
* **Secondary Data**: Collected and published by third parties (e.g., Kaggle datasets or institutional databases).

### Collection Techniques
1. **Registration**: Recording administrative data from specific institutions.
2. **Sensus**: Investigating every element in a population for maximum accuracy.
3. **Survey (Sampling)**: Investigating a representative sample to represent a population.
4. **Experiment**: Manipulating conditions to observe specific effects or influences.
5. **Observation**: Recording data directly without intervention or manipulation.

```python
# Test Code 2: Web Scraping using BeautifulSoup
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import pytz

# Target URL for the scraping exercise
url = 'https://www.kompas.com/food/read/2026/03/08/153100575/resep-craquelin-klepon-jajanan-fusion-indonesia-perancis-untuk-lebaran'

# Send GET request
response = requests.get(url)

if response.status_code == 200:
    # Initialize parser
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract Article Title
    title = soup.find('h1').text.strip()

    # Extract and format publication date
    date_meta = soup.find('meta', property='article:published_time')
    if date_meta:
        # Convert UTC to Asia/Jakarta (WIB)
        utc_time = datetime.fromisoformat(date_meta['content'].replace('Z', '+00:00'))
        wib_time = utc_time.astimezone(pytz.timezone('Asia/Jakarta'))
        pub_date = wib_time.strftime('%d/%m/%Y, %H:%M WIB')
    else:
        pub_date = "Not found"

    # Extract Tags/Keywords
    tags_meta = soup.find('meta', {'name': 'keywords'})
    tags = tags_meta['content'] if tags_meta else "Not found"

    print(f"Title    : {title}")
    print(f"Date     : {pub_date}")
    print(f"Tags     : {tags}")
else:
    print(f"Failed to access URL. Status code: {response.status_code}")
```

## Advanced Data Acquisition (Kaggle)

Kaggle is a major platform for secondary data acquisition. For this case study, we use a banking dataset to analyze customer patterns.

### Automated Kaggle Integration
Instead of manual downloads, we use the `kagglehub` library to fetch datasets directly into the environment.

```python
# Downloading Kaggle Dataset
import kagglehub
import pandas as pd
import os

# Case Study: Bank Customer Churn Prediction
# Handle: adammaus/predicting-churn-for-bank-customers
try:
    path = kagglehub.dataset_download("adammaus/predicting-churn-for-bank-customers")
    print("Dataset location:", path)

    # Identify the CSV file
    files = os.listdir(path)
    csv_file = [f for f in files if f.endswith('.csv')][0]

    # Load into DataFrame
    df_bank = pd.read_csv(os.path.join(path, csv_file))
    print("\nKaggle Dataset Preview (Top 5 Rows):")
    print(df_bank.head(5))
except Exception as e:
    print(f"Kaggle download error: {e}")
```
