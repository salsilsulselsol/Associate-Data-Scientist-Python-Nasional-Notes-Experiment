# Notebook 11: Data Annotation Implementation
**Dataset:** Banking Credit Risk (Synthetic)

**Task:** Annotate customers into Risk Categories based on Credit Scores.

### Annotation SOP (Standard Operating Procedure):
1. **High Priority (Prime)**: Credit Score > 750 (Low risk, high approval priority).
2. **Medium Priority (Subprime)**: Credit Score between 600 - 750 (Requires manual review).
3. **Low Priority (High Risk)**: Credit Score < 600 (Likely rejection).

## Section 1: Data Preparation and Annotation logic

```python
import pandas as pd
import numpy as np

# Creating a synthetic dataset for practice
data = {
    'customer_id': range(101, 111),
    'credit_score': [800, 450, 720, 680, 300, 850, 610, 590, 760, 400],
    'income_level': ['High', 'Low', 'Medium', 'Medium', 'Low', 'High', 'Medium', 'Low', 'High', 'Low']
}

df = pd.DataFrame(data)
```

## Section 2: Implementing Annotation Based on SOP

```python
# Define a function to apply SOP rules
def apply_credit_sop(score):
    if score > 750:
        return 'Prime'
    elif 600 <= score <= 750:
        return 'Subprime'
    else:
        return 'High Risk'

# Creating the 'target_label' column as per SOP
df['target_label'] = df['credit_score'].apply(apply_credit_sop)

# Section 3: Result Verification
print("--- Annotated Dataset Based on Banking SOP ---")
print(df[['customer_id', 'credit_score', 'target_label']])

# Checking distribution for validation
print("\n--- Annotation Distribution ---")
print(df['target_label'].value_counts())
```

### Analysis of Labeling Results:
* **SOP Compliance**: The labeling process strictly follows the defined thresholds (>750, 600-750, <600).
* **Pattern Identified**: Customers with 'High' income levels consistently fall into the 'Prime' category, showing a strong correlation between features and labels.
* **Model Readiness**: The `target_label` column is now encoded and ready for a Multi-class Classification task.
